```python
import reverse_geocoder as rg, pandas as pd
import geocoder
stations=pd.read_csv('..\ghcnm_out\Correlated Data\Monthly_Correlated_Data.csv')
iceland=pd.DataFrame({'Code':[], 'Latitude':[], 'Longitude':[], 'Elevation':[], 'Name':[]})
```

```python
print(len(stations))
for i in range(0, len(stations)):
    print(i)
    g = geocoder.osm([stations.iloc[i,1], stations.iloc[i,2]], method='reverse')
    print(g.json)
    if (g.json['country_code']=='is' and g.json['country']=='Ísland'):
        iceland=iceland.append(stations.iloc[i])
print(iceland)
```

```python
iceland.to_csv('Iceland_stations.csv')
```
