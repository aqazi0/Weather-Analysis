```python
import os
import pandas as pd
import numpy as np
import plotly.express as px
import plotly
```

```python
df=pd.read_csv('ghcnm_out/Correlated Data/Monthly_Correlated_Data.csv')
```

```python
for i in range(len(df)):
    if(df.iloc[i,6]>0.65 or df.iloc[i,6]<-0.65):
        mon1=[]
        mon2=[]
        code=df.iloc[i,1]
        data=pd.read_csv('ghcnm_out/data/'+code+'.csv')
        for j in range(len(data)):
            if(data.iloc[j,0][-2:]=='01'):
                if(pd.notna(data.iloc[j,1]) and pd.notna(data.iloc[j+1,1])):
                    mon1.append(data.iloc[j,1])
                    mon2.append(data.iloc[j+1,1])
        lst=list(zip(mon1,mon2))
        plotdata = pd.DataFrame(lst,columns = ['Jan_Temp', 'Feb_Temp'])
        title_=df.iloc[i,1]+':'+str(df.iloc[i,6])
        fig = px.scatter(plotdata, x="Jan_Temp", y="Feb_Temp", trendline="ols", title=title_)
        if not os.path.exists('ScatterGrams\\'+code):
            os.mkdir('ScatterGrams\\'+code)
        plotly.offline.plot(fig, filename='ScatterGrams\\'+code+'\\1.JanFeb.html')
        fig.write_image('ScatterGrams\\'+code+'\\1.JanFeb.png')
```
